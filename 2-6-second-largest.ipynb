```python
class Secondlargest:
    def second_largest(self, nums: list[int]) -> int | None:
        if not isinstance(nums,list):
            print('Please provide a list')
            return None
        
        if len(nums) < 2:
            return None
        
        first_max = float('-inf')
        second_max = float('-inf')

        for num in nums:
            if num > first_max:
                second_max = first_max
                first_max = num
            elif first_max > num > second_max:
                second_max = num

        return second_max

obj1 = Secondlargest()
obj1.second_largest([1,4,66,8,111,7,99,211,107])
```
