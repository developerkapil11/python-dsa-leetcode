```python
class RemoveElement:
    def removeElement(self, nums:list[int], val:int):
        x = 0
        for i in range(len(nums)):
            if nums[i] != val:
                nums[x] = nums[i]
                x += 1
        result = nums[:x] + ["_"] * (len(nums) - x)
        return result, x

obj = RemoveElement()
print(obj.removeElement([1,1,0,2,2,4,6,2,5], 2))
```
