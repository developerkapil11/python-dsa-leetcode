```python
class Reverseinteger:
    def reverse_integer(self, nums:int) -> int:
        sign = -1 if nums < 0 else +1
        nums = abs(nums)
        rev = 0
        while nums > 0:

            rev = rev * 10 + nums % 10
            nums = nums//10

        rev *= sign
        return rev
    
obj = Reverseinteger()
print(obj.reverse_integer(12345678))
print(obj.reverse_integer(7846544783))
print(obj.reverse_integer(-354644))
```
