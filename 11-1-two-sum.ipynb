```python
class Twosum:
    def two_sum(self, nums: list[int], target:int) -> list[int]:
        if not isinstance(nums, list):
            print("Please Provide an List")
            return []
        
        obj = {}
        for i in range(len(nums)):
            val = target - nums[i]
            if val in obj:
                return [obj[val], i]
            obj[nums[i]] = i
        return []
    
sum1 = Twosum()
sum1.two_sum([2,7,11,25], 9)
```
