```python
class Countdigit:
    def countdigit(self, nums: int) -> int | None:
        nums = abs(nums) #number converted in +ve if nsomeone count -ve number

        if not isinstance(nums, int):
            return None
        
        if nums == 0: return 1
        count = 0

        while nums > 0:
            nums = nums//10
            count+= 1
        
        return count

obj = Countdigit()
print(obj.countdigit(2347485777934729255))
print(obj.countdigit(0))
print(obj.countdigit(-12345))
print(obj.countdigit(999999999))
print(obj.countdigit(123))   # None
```
