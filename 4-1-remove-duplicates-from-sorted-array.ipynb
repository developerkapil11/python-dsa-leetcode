```python
class RemoveDuplicate:
    def removeDuplicate(self, nums: list[int]) -> int:
        x = 0
        for i in range(len(nums)):
            if nums[i] > nums[x]:
                x += 1
                nums[x] = nums[i]
        return x+1

obj = RemoveDuplicate()

print(obj.removeDuplicate([0,0,1,2,2,3,4,4,5,5,6,6,6,7,8,8,9]))
print(obj.removeDuplicate([1]))
print(obj.removeDuplicate([1,1,2,2,2,3,4,5,5,5,5]))
print(obj.removeDuplicate([5,5,6,6]))
```
